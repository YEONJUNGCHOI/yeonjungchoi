```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np 
%matplotlib inline
```

```python

def AND(x1 , x2):
    w1 , w2 , theta = 0.5 , 0.5 , 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0;
    elif tmp>theta:
        return 1
```

```python
print AND(0,0)
print AND(1,0)
print AND(0,1)
print AND(1,1)
```

```python
import numpy as np
def NAND(x1 ,x2):
    x = np.array([x1, x2])
    w = np.array([-0.5 , -0.5])
    b = 0.7
    
    tmp = np.sum(w*x)+b
    if tmp <=0:
        return 0
    else:
        return 1
```

```python
def OR(x1 , x2):
    x = np.array([x1 , x2])
    w = np.array([ 0.5 , 0.5])
    b = -0.2
    tmp = np.sum(x*w) +b
    if tmp<= 0:
        return 0
    else:
        return 1
```

```python
print NAND(0,0)
print NAND(0,1)
print NAND(1,0)
print NAND(1,1)
```

```python
def XOR(x1,x2):
    s1=NAND(x1,x2)
    s2=OR(x1,x2)
    y=AND(s1,s2)
    return y
```

```python
print XOR(0,0)
print XOR(1,0)
print XOR(0,1)
print XOR(1,1)
```

```python
def step(x):
    y=x>0
    y=y.astype(np.int)
    return y
```

```python
x=np.arange(-2.0 , 2.0 ,0.001)
y=step(x)
plt.ylim(-2,2)
plt.plot(x,y)
```

```python
def Sigmoid(x):
    return 1 / (np.exp(-x)+1)
```

```python
def exponential(x):
    return np.exp(x)
```

```python
x=np.arange(-2 , 2 , 0.0001)
y=exponential(x)
plt.plot(x,y)
plt.title('exponential')
```

```python
def minus_exponential(x):
    return np.exp(x)*-1
```

```python
x=np.arange(-2 , 2 , 0.0001)
y=minus_exponential(x)
plt.plot(x,y)
plt.title('exponential')
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-5 , 5 , 0.0001)
y=relu(x)
plt.plot(x,y)
plt.ylim(-2,2)
plt.xlim(-2,2)
plt.title('Relu')
```

```python
def minus_exponential_rescipocal(x):
    return 1/np.exp(x)*-1
```

```python
x=np.arange(-2 , 2 , 0.0001)
y=minus_exponential_rescipocal(x)
plt.plot(x,y)
plt.title('minus_exponential_rescipocal')
```

```python
x=np.arange(-2 , 2 , 0.0001)
y=Sigmoid(x)
plt.plot(x,y)
plt.title('sigmoid')
```
